```python
%%cython 
import cython
import numpy as np
cimport numpy as np

# DTYPE = np.float64
# ctypedef np.float64_t DTYPE_t
DTYPE = np.int64
ctypedef np.int_t DTYPE_t

@cython.boundscheck(False)
@cython.wraparound(False)
def mmax(np.ndarray[DTYPE_t, ndim=2] A):
    cdef int i,j
    cdef DTYPE_t Amax
    cdef int a1 = A.shape[0]
    cdef int a2 = A.shape[1]
    
    for i in range(0,a1):
        for j in range(0,a2):
            if i==0 and j==0:
                Amax = A[i,j]
            elif A[i,j] > Amax:
                Amax = A[i,j]
                
    return Amax    
```
